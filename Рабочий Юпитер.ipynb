```python
data["MKB1_id"]=pd.factorize(data.MKB1)[0]+1
data["MKB2_id"]=pd.factorize(data.MKB2)[0]+1
data["PMU_id"]=pd.factorize(data.PMU)[0]+1
data["Crit_id"]=pd.factorize(data.Crit)[0]+1
data.fillna(0, inplace=True)
data
```

```python
mkb1_values=data[['MKB1',"MKB1_id"]].copy()
mkb1_values=mkb1_values.drop_duplicates(subset=['MKB1',"MKB1_id"])
mkb2_values=data[['MKB2',"MKB2_id"]].copy()
mkb2_values=mkb2_values.drop_duplicates(subset=['MKB2',"MKB2_id"])
crit_values=data[['Crit','Crit_id']].copy()
crit_values=crit_values.drop_duplicates(subset=['Crit','Crit_id'])
Pmu_values=data[['PMU','PMU_id']].copy()
Pmu_values=Pmu_values.drop_duplicates(subset=['PMU','PMU_id'])
```

```python
crit_values
```

```python
mkb1_values.to_csv("mkb1_values.csv", index=False)
mkb2_values.to_csv("mkb2_values.csv", index=False)
crit_values.to_csv("crit_values.csv", index=False)
Pmu_values.to_csv("Pmu_values.csv", index=False)
```

```python
X = pd.DataFrame(data, columns=[ 'Days', 'Prof', 'MKB1_id', 'MKB2_id', 'Sex', 'Age', 'Month', 
       'PMU_id', 'Crit_id', 'Result', 'Paysum'])
y = pd.DataFrame(data, columns=['Exp'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
n_estimators = [600,700]
max_features = ['sqrt']
max_depth = [20,30,70,110,150]
min_samples_split = [2,3,4,22,23,40]
min_samples_leaf = [2,3,4,5,6,7]
bootstrap = [False]
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
gs = GridSearchCV(RandomForestClassifier(), param_grid, cv = 3, verbose = 1, n_jobs=-1)
gs.fit(X_train, y_train)
model = gs.best_estimator_
gs.best_params_
```

```python
y_pred_gs = gs.best_estimator_.predict(X_test)
```

```python
best={'bootstrap': False,
 'max_depth': 150,
 'max_features': 'sqrt',
 'min_samples_leaf': 2,
 'min_samples_split': 22,
 'n_estimators': 700}
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_gs)
r2
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred_gs ))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(bootstrap= False,
 max_depth= 150,
 max_features= 'sqrt',
 min_samples_leaf= 2,
 min_samples_split= 22,
 n_estimators= 700)
model.fit(X_train, y_train)
```

```python
y_pred_model = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred_model  ))
```

```python
import joblib
joblib.dump(model, "model.joblib", compress=9) 
```

```python
loaded_model = joblib.load( "model.joblib")
```

```python
y_pred_loaded_model = loaded_model.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred_loaded_model  ))
```
